```python
import pandas as pd
import os 
```

```python
amazon_sample_products = pd.read_gbq(

    query="""

        SELECT

            product_asin,

            variation_asin,

            NAME,

            DESCRIPTION,

            MAIN_IMAGE

        FROM

            `brain-ext-shop-data-dev.zyte_amazon_data_dwh.dist_prediction_product` 

        WHERE TRUE

            AND LOWER(BREADCRUMBS_CATEGORY_NAME) LIKE "%bekleidung%"

            AND NAME IS NOT NULL

            AND DESCRIPTION IS NOT NULL

            AND MAIN_IMAGE IS NOT NULL

            AND product_asin IS NOT NULL

            AND VARIATION_ASIN IS NOT NULL

        LIMIT 10

  """,

    use_bqstorage_api=True,

    project_id="brain-sonar-dev",

)
```
